#### Ejercicio 01. Análisis de datos básico con estructuras nativas de python.
Crea una lista con las calificaciones de 5 estudiantes: [8.5, 9.2, 7.8, 8.9, 10].

- Calcula el promedio de las calificaciones.

```python
# Código
calificaciones = [8.5,9.2,7.8,8.9,10]

promedio = sum(calificaciones) / len(calificaciones)
print("promedio de calificaciones", promedio)
```

- Encuentra la calificación más alta y la más baja.

```python
# Código
calificacion_mas_alta = max(calificaciones)
calificacion_mas_baja = min(calificaciones) 
print("calificacion mas alta", calificacion_mas_alta)
print("calificacion mas baja", calificacion_mas_baja)
```

#### Ejercicio 01. ¿Cuál es la casa más cara en todo el dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle del Prado, Nº20 es la más cara y su precio es de 5000000 USD`

#### Ejercicio 05. ¿El dataset contiene valores nulos (NAs)? (★☆☆)

Imprima un booleano (`True` o `False`) para verificar si hay valores nulos seguido de las columnas que contiene el NAs.

```python
# TODO
# verificar si hay valores nulos 
tiene_valores_nulos = df.isnull().any().any()

# encontrar columnas con valores nulos 

columnas_nulas = df.columns[df.isnull().any()]

print("elconjunto de datos tiene valores nulos", tiene_valores_nulos)
print("columnas con valores nulos", columnas_nulas.tolist())
```

#### Ejercicio 06. Elimina los valores nulos (NAs) del dataset, si aplica (★★☆)

Despues de eliminar los valores nulos, compara el tamaño del DataFrame antes y después de la eliminación.


```python
# TODO
# tama;o original 
original_size = df.shape 

# eliminar filas con valor nulo 
df_cleaned = df.dropna()

# tamaño de los datos despues de limpiar 

cleaned_size = df_cleaned.shape 

# comparar tamaños 

print("tamaño original de los datos", original_size)
print("datos limpios", cleaned_size)
print("filas eliminadas:", original_size[0] - cleaned_size[0])
```

#### Ejercicio 07. ¿Cuál es la media de precios en la población de "Arroyomolinos (Madrid)"? (★★☆)

Imprima el valor obtenido de la columna level5.

```python
# TODO
# filtrar los datos para la poblacion especificada
arroyomolinos_df = df[df["level5"] == "arroyomolinos (madrid)"]

# calcular el precio promedio
precio_promedio_arroyomolinos = arroyomolinos_df["price"].mean()

# imprimir el precio promedio
print("precio promedio en arroyomolinos (madrid):", precio_promedio_arroyomolinos)
```

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt

# Trazar el histograma de precios
plt.hist(arroyomolinos_df["price"], bins=20)
plt.xlabel("precio")
plt.ylabel("frecuencia")
plt.title("histograma de precios en arroyomolinos (madrid)")

print ("el histograma muestra la distribucion de los precios de las casas en arroyomolinos (madrid). \n"
     "se observa una distribucion hacia la derecha, lo que indica que hay mas casas con precios bajos y menos casas con precios altos. ")
```
